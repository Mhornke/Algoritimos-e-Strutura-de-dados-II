```python
def GeraConjuntosBooleanos(lista_perguntas):
    if not lista_perguntas:
        return [[]]

    primeiro = lista_perguntas[0]
    resto = lista_perguntas[1:]

    combinacoes_resto = GeraConjuntosBooleanos(resto)

    resultado = []

    for combinacao in combinacoes_resto:
        tupla_false = (primeiro, False)
        resultado.append([tupla_false] + combinacao)

        tupla_true = (primeiro, True)
        resultado.append([tupla_true] + combinacao)

    return resultado


```

```python
todas = GeraConjuntosBooleanos(condicoes)
print(f"Total de casos gerados: {len(todas)}")
print("\nCasos Gerados")

for i, c in enumerate(todas):
  print(f"\nCasos","[", i ,"]", c,"\n")

  if c[0][1] == True:
    print("Clinte é maior de idade")
  else:
    print("Cliente não é maior de idade")

  if c[1][1] == True:
    print("Clinte Tem conta ativa")
  else:
    print("Clinte não Tem conta ativa")

  if c[2][1] == True:
        print("Clinte Tem credito")
  else:
        print("Clinte não Tem credito")

  if c[3][1] == True:
    print("Cliente premium")
  else:
      print("Cliente não premium")






    # print(f"Caso","[",i,"]", "\n"condicao[0], "\nMaior de idade: ", cliente['idade'] >= 18, "\nPossui conta: ", cliente['conta'].strip() == '', "\nConta Premium: ", cliente['premium'], "\n")
```
