```python
# Q2

# Você tem um array com 100 valores aleatórios e quer saber se o número 25 está lá.

# Como os valores são aleatórios (não ordenados), você precisa verificar um por um.
# No pior caso, o 25 pode estar no final ou nem existir.

# Complexidade no pior caso:
# O(n)
# ele é um O(n) por que precisa percorrer todos os elementos para achar o 25.
# Porque você pode precisar olhar todos os elementos.
```

```python
# Q3

# Por que acessar por índice é mais rápido que buscar um valor?

# Por que procurar um valor dentro de um array não ordenado exige ums busca lienar
# e comparando cada valor com o elemento desejado já a busca por indice é mais rapida
# por ir direto no indice do valor que esta armazenado na memoria

# Acesso por índice: vai direto na posição → não precisa procurar
# Busca por valor: precisa comparar elemento por elemento até encontrar

# Resumindo:

# Índice = acesso direto (tipo endereço)
# Busca = varrer a lista inteira

# Por isso:

# Índice → O(1)
# Busca → O(n)
```
