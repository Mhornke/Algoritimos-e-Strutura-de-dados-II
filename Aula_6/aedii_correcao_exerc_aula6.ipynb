Classe ListaLigadaCompleta

```python
class No:
    def __init__(self, dado):
        self.dado = dado
        self.proximo = None

class ListaLigadaCompleta:
    def __init__(self):
        self.head = None
        self.tail = None

    def inserir_fim(self, dado):
        novo = No(dado)
        if self.tail is None:
            self.head = self.tail = novo
        else:
            self.tail.proximo = novo
            self.tail = novo

    def inserir_inicio(self, dado):
        novo = No(dado)
        if self.head is None:
            self.head = self.tail = novo
        else:
            novo.proximo = self.head
            self.head = novo

    def remover_inicio(self):
        if self.head is None:
            print("Lista vazia.")
            return None
        removido = self.head.dado
        self.head = self.head.proximo
        if self.head is None:
            self.tail = None
        return removido

    def remover_valor(self, dado):
        atual = self.head
        anterior = None
        while atual:
            if atual.dado == dado:
                if anterior is None:
                    self.head = atual.proximo
                else:
                    anterior.proximo = atual.proximo
                if atual.proximo is None:
                    self.tail = anterior
                return True
            anterior = atual
            atual = atual.proximo
        print(f"'{dado}' não encontrado.")
        return False

    def buscar(self, dado):
        atual = self.head
        while atual:
            if atual.dado == dado:
                return True
            atual = atual.proximo
        return False

    def exibir(self):
        elementos = []
        atual = self.head
        while atual:
            elementos.append(atual.dado)
            atual = atual.proximo
        print(elementos)
```

Parte 1.1: Inicialização com Lista Ligada

```python
lista = ListaLigadaCompleta()
pacientes = ['Ana', 'Bruno', 'Carla', 'Daniel', 'Eduarda', 'Felipe']

for paciente in pacientes:
    lista.inserir_fim(paciente)

print('Lista Ligada inicial:')
lista.exibir()
```

Parte 1.2: Inicialização com Array

```python
array = []
pacientes = ['Ana', 'Bruno', 'Carla', 'Daniel', 'Eduarda', 'Felipe']

for paciente in pacientes:
    array.append(paciente)

print('Array inicial:')
print(array)
```

Parte 2.1: Operações com Lista Ligada

```python
# 1. Inserir 'Joao' no inicio
lista.inserir_inicio('Joao')

# 2. Inserir 'Marina' no fim
lista.inserir_fim('Marina')

# 3. Verificar se 'Carla' esta na lista
if lista.buscar('Carla'):
    print('Carla encontrada na Lista Ligada')
else:
    print('Carla NAO encontrada')

# 4. Remover o primeiro elemento
removido = lista.remover_inicio()
print(f"Primeiro elemento removido: '{removido}'")

# 5. Remover 'Daniel' pelo valor
lista.remover_valor('Daniel')

# 6. Exibir estrutura final
print('Lista Ligada final:')
lista.exibir()
```

Parte 2.2: Operações com Array

```python
# 1. Inserir 'Joao' no inicio
array.insert(0, 'Joao')

# 2. Inserir 'Marina' no fim
array.append('Marina')

# 3. Verificar se 'Carla' esta na lista
if 'Carla' in array:
    print('Carla encontrada no Array')
else:
    print('Carla NAO encontrada')

# 4. Remover o primeiro elemento
removido = array.pop(0)
print(f"Primeiro elemento removido: '{removido}'")

# 5. Remover 'Daniel' pelo valor
array.remove('Daniel')

# 6. Exibir estrutura final
print('Array final:')
print(array)
```

Parte 3: Respostas das Questões

Questão 1 — Inserção no Início
A Lista Ligada é mais vantajosa para inserção no início.
Na Lista Ligada, basta criar um novo nó e apontar seu 'proximo' para o
head atual, depois atualizar o head, operação O(1), independente do tamanho.
No Array, inserir no início exige deslocar todos os elementos uma posição
para a direita, resultando em O(n).


Questão 2 — Busca por Valor
Ambas as estruturas têm comportamento semelhante na busca por valor: O(n).
Isso ocorre porque nenhuma das duas mantém os dados ordenados ou indexados
por valor. Em ambos os casos é necessário percorrer elemento por elemento
até encontrar (ou não) o valor desejado.


Questão 3 — Lista Ligada é boa escolha para a Clínica?
Sim, a Lista Ligada é uma boa escolha para a clínica, especialmente por conta
das operações mais frequentes:
- Inserção no início (paciente prioritário): O(1) na Lista Ligada vs O(n) no Array.
- Remoção no início (atender o próximo): O(1) na Lista Ligada vs O(n) no Array.
- Remoção por nome (desistência): O(n) em ambas, empate.
Como a fila prioriza inserções e remoções nas extremidades, a Lista Ligada
leva vantagem clara sobre o Array nesse cenário.


Questão 4 — Quando o Array é melhor que a Lista Ligada?
O Array é melhor quando:
- É necessário acesso direto por índice: O(1) no Array vs O(n) na Lista Ligada.
- As inserções e remoções ocorrem predominantemente no fim da estrutura.
- A memória disponível é limitada (o Array não carrega ponteiros extras).
- É necessária melhor localidade de cache (dados contíguos na memória).
Exemplos: tabelas de consulta, histórico de atendimentos por número de ordem,
ou quando se percorre a lista com frequência por posição.
