```python
valores = [10,5,2,7,15,12,20]
```

```python
class Node():
  def __init__(self, valor):
    self.valor = valor
    self.esq = None
    self.dir = None
```

```python
print(
  Node(valores[0]).valor,
  Node(valores[0]).esq,
  Node(valores[0]).dir)
```

```python
def inserir(raiz, valor):
  if raiz is None:
    return Node(valor)
  if valor < raiz.valor:
    raiz.esq = inserir(raiz.esq, valor)
  else:
    raiz.dir = inserir(raiz.dir, valor)
  return raiz
```

```python
# Testando para ver como funciona o `inserir`
tree = inserir(None, valores[0])
print(
    tree.valor,
    tree.esq,
    tree.dir
)
```

```python
tree = None

# [10,5,2,7,15,12,20]
for valor in valores:
    tree = inserir(tree, valor)

print(f"Raiz: {tree.valor}, Valor da Esquerda da Raiz {tree.esq.valor}, Valor da Direita da Raiz {tree.dir.valor}")
```
