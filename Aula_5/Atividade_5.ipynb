```python
# 1. Inserção em O(n): Crie uma função inserirNoInicio(vetor, novoPaciente). Você deve
# percorrer o vetor do fim para o começo, movendo cada elemento uma posição para
# a direita, e então inserir o novo paciente no índice 0.
# *apesar de não remover um item especifico ele tira o ultimo item da lista*

indice = 0
def inserirNoInicio(vetor, novoPaciente):

  movi = 0
  for i in range(len(vetor) - 1, indice, -1):
    vetor[i] = vetor[i - 1]
    movi += 1
  vetor[indice] = novoPaciente

  return(movi, vetor)



def inserirNoFim(vetor, novoPaciente):
  movi = 0

  tamanho = len(vetor) -1
  vetor[tamanho] = novoPaciente
  movi += 1

  return(movi,vetor)
  print(vetor)


```

```python
# 2. Remoção com Compactação: Crie uma função removerDaPosicao(vetor, indice).
# Após apagar o dado, mova todos os elementos à direita para a esquerda, garantindo
# que não fiquem "buracos" (valores nulos/undefined) no meio do vetor.

def removerDaPosicao(vetor, indice):


  tamanho = len(vetor) -1
  movi = 0

  for i in range(indice, len(vetor) -1, +1):
    movi += 1
    vetor[i] = vetor[i + 1]

  vetor[tamanho] = None
  print(vetor)
  print(f"Vezes que movimentou os elementos: {movi}")


removerDaPosicao(vetor, 0)
```

```python
# BLOCO 2
# 4. Acesso Direto: Demonstre a eficiência O(1). Crie uma função que receba o vetor e
# retorne a temperatura da "hora agendada" pelo usuário sem percorrer o array.
import random
c = [random.randint(20,36) for i in range(24)]

def horaAgendada(c, hora):

  busca = c[hora]
  print(busca)
  return busca



# 5. Busca Manual: Implemente uma função existeTemperatura(vetor, valorBuscado).
# Use um laço para percorrer o vetor e retorne o índice da primeira ocorrência. Se não
# encontrar, retorne -1.

def existeTemperatura(vetor, valorBuscado):
  # busca = [(i, valorVetor) for i, valorVetor in enumerate(vetor) if  valorVetor == valorBuscado ]
  indice = 0
  retorno = 0
  for temperatura in vetor:
    isTemperatura = temperatura == valorBuscado
    indice += 1


    if indice == len(vetor) -1 and retorno == 0:

      return -1, indice
    elif isTemperatura:

      return 1, indice





# 6. Análise de Extremos: Implemente uma função que encontre o maior e o menor
# valor do vetor percorrendo-o apenas uma vez (Complexidade O(n)).

def maiorMenor(vetor):
  controle = 1
  maior = vetor[0]
  menor = vetor[0]

  while controle <= len(vetor) -1:
    item = vetor[controle]

    comparaMenor = menor > item
    comparaMaior = maior < item

    if comparaMenor:
        menor = item

    if comparaMaior:
        maior = item

    controle += 1
  return maior, menor


q_quatro = horaAgendada(c, 20)
q_cinco, indice = existeTemperatura(c, 30)
q_seisMaior, q_seisMenor = maiorMenor(c)

print(f"vetor: {c}")
print(f"Quetão 4: {q_quatro}")
print(f"Quetão 5: {q_cinco}, indice: {indice}")
print(f"Quetão 6: Maior Temp: {q_seisMaior}, Menor Temp: {q_seisMenor}")
```
